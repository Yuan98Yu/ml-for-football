```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.ion()

import torch
from torch.utils.data import DataLoader
from torchvision import transforms, utils

from FootballDataset import FootballDataset
from FootballDisplay import FootballDisplay
from Transforms import *
```

```python
train_dataset = FootballDataset(train=True, 
                                transform=transforms.Compose([ToTensor(),
                                                            YTo2D()]))
test_dataset = FootballDataset(train=False,
                               transform=transforms.Compose([ToTensor(),
                                                            YTo2D()]))
```

```python
print(train_dataset[:]['X'].dtype)
```

```python
print(train_dataset[:]['X'].shape, train_dataset[:]['Y'].shape)
print(test_dataset[:]['X'].shape, test_dataset[:]['Y'].shape)
```

```python
train_displayer = FootballDisplay(train_dataset)
test_displayer = FootballDisplay(test_dataset)
```

```python
train_displayer.show_football(1500, 30)
test_displayer.show_football(500, 30)
plt.show()
```

```python
train_displayer.save_gifs(range(3), 'train')
test_displayer.save_gifs(range(3), 'test')
```

```python
train_loader = DataLoader(train_dataset, batch_size=4,
                        shuffle=True, num_workers=4)
test_loader = DataLoader(test_dataset, batch_size=4,
                        shuffle=False, num_workers=4)
```

```python
for i_batch, sample_batched in enumerate(train_loader):
    print(i_batch, sample_batched['X'].size(),
          sample_batched['Y'].size())

    # observe 4th batch and stop.
    if i_batch == 3:
        break

for i_batch, sample_batched in enumerate(test_loader):
    print(i_batch, sample_batched['X'].size(),
          sample_batched['Y'].size())

    # observe 4th batch and stop.
    if i_batch == 3:
        break
```
